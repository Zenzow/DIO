```python
Brazil = df.loc[df["country"]=="Brazil"]
Brazil
```

```python
Brazil.groupby(["year"]).sum()
```

```python
#casos de covid no brasil em 2020 / semana
plt.figure(figsize=(6, 12))
ax1 = plt.subplot(211)
plt.title("Casos de Covid no brasil em 2020 por semana", fontdict = {'fontsize': 16})
Brazil_cases_2020 = Brazil.loc[(Brazil["indicator"] == "cases") & (Brazil["year"] == 2020)]
Brazil_cases_2020.plot.line(x="week", y="weekly_count", ax=ax1)
```

```python
#casos de covid no brasil em 2021 / semana
plt.figure(figsize=(6, 12))
ax1 = plt.subplot(211)
plt.title("Casos de Covid no brasil em 2021 por semana", fontdict = {'fontsize': 16})
Brazil_cases_2021 = Brazil.loc[(Brazil["indicator"] == "cases") & (Brazil["year"] == 2021)]
Brazil_cases_2021.plot.line(x="week", y="weekly_count", ax=ax1)
```

```python
#mortes de covid no brasil em 2020 / semana
plt.figure(figsize=(6, 12))
ax1 = plt.subplot(211)
plt.title("Mortes de Covid no brasil em 2020 por semana", fontdict = {'fontsize': 16})
Brazil_cases_2020 = Brazil.loc[(Brazil["indicator"] == "deaths") & (Brazil["year"] == 2020)]
Brazil_cases_2020.plot.line(x="week", y="weekly_count", ax=ax1)
```

```python
#mortes de covid no brasil em 2021 / semana
plt.figure(figsize=(6, 12))
ax1 = plt.subplot(211)
plt.title("Mortes de Covid no brasil em 2021 por semana", fontdict = {'fontsize': 16})
Brazil_cases_2020 = Brazil.loc[(Brazil["indicator"] == "deaths") & (Brazil["year"] == 2021)]
Brazil_cases_2020.plot.line(x="week", y="weekly_count", ax=ax1)
```

```python
#casos por continente em 2020
df_cases_2020 = df.loc[(df["indicator"] == "cases") & (df["year"] == 2020)]
df_continent = df_cases_2020.groupby("continent")

df_continent["weekly_count"].mean()
```

```python
#casos por continente em 2021
df_cases_2021 = df.loc[(df["indicator"] == "cases") & (df["year"] == 2021)]
df_continent = df_cases_2021.groupby("continent")

df_continent["weekly_count"].mean()
```

```python
#mortes por continente em 2020
df_deaths_2020 = df.loc[(df["indicator"] == "deaths") & (df["year"] == 2020)]
df_continent = df_deaths_2020.groupby("continent")

df_continent["weekly_count"].mean()
```

```python
#casos por continente em 2021
df_deaths_2021 = df.loc[(df["indicator"] == "deaths") & (df["year"] == 2021)]
df_continent = df_deaths_2021.groupby("continent")

df_continent["weekly_count"].mean()
```

```python
#progressão casos de covid no mundo em 2020 / semana
plt.figure(figsize=(6, 20))
ax1 = plt.subplot(211)
plt.title("Casos de Covid no mundo em 2020 por semana", fontdict = {'fontsize': 16})
df_cases_2020 = df.groupby(["week","year","indicator"]).sum()
df_cases_2020 = df_cases_2020.reset_index()
df_cases_2020
df_cases_summed_2020 = df_cases_2020.loc[(df_cases_2020["indicator"] == "cases") & (df_cases_2020["year"] == 2020)]
df_cases_summed_2020
df_cases_summed_2020.plot.line(x="week", y="weekly_count", ax=ax1)
```
